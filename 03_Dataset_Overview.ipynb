## 6. Cross-File Relationships

The NorthStar datasets are connected through shared identifiers. These relationships allow operational records from different business functions to be integrated for later SQL analysis, customer complaint investigation, delivery performance analysis, and MongoDB document modelling.

The main relationships are:

- `customers.customer_id` → `orders.customer_id`
- `customers.customer_id` → `complaints.customer_id`
- `customers.customer_id` → `app_events.customer_id`
- `orders.order_id` → `deliveries.order_id`
- `orders.order_id` → `complaints.order_id`
- `orders.order_id` → `app_events.order_id`
- `deliveries.delivery_id` → `incidents.delivery_id`
- `deliveries.driver_id` → `drivers.driver_id`
- `deliveries.vehicle_id` → `vehicles.vehicle_id`
- `deliveries.hub_id` → `hubs.hub_id`

```python
# Verify key relationships between the main NorthStar datasets
relationship_check = pd.DataFrame({
    "Relationship": [
        "Orders linked to Customers",
        "Complaints linked to Customers",
        "App Events linked to Customers",
        "Deliveries linked to Orders",
        "Complaints linked to Orders",
        "App Events linked to Orders",
        "Incidents linked to Deliveries",
        "Deliveries linked to Drivers",
        "Deliveries linked to Vehicles",
        "Deliveries linked to Hubs"
    ],
    "Matched Records": [
        orders["customer_id"].isin(customers["customer_id"]).sum(),
        complaints["customer_id"].isin(customers["customer_id"]).sum(),
        app_events["customer_id"].isin(customers["customer_id"]).sum(),
        deliveries["order_id"].isin(orders["order_id"]).sum(),
        complaints["order_id"].isin(orders["order_id"]).sum(),
        app_events["order_id"].isin(orders["order_id"]).sum(),
        incidents["delivery_id"].isin(deliveries["delivery_id"]).sum(),
        deliveries["driver_id"].isin(drivers["driver_id"]).sum(),
        deliveries["vehicle_id"].isin(vehicles["vehicle_id"]).sum(),
        deliveries["hub_id"].isin(hubs["hub_id"]).sum()
    ],
    "Total Records": [
        len(orders),
        len(complaints),
        len(app_events),
        len(deliveries),
        len(complaints),
        len(app_events),
        len(incidents),
        len(deliveries),
        len(deliveries),
        len(deliveries)
    ]
})

relationship_check
```

## 8. Summary of Dataset Overview

The dataset overview confirmed that the NorthStar project contains multiple connected operational files covering customers, orders, deliveries, complaints, drivers, hubs, vehicles, incidents, and app events. The dataset includes both structured relational records and flexible event-based data, making it suitable for combined SQL, R, Python, and MongoDB analysis.

The initial inspection identified several important data-quality issues, including missing values in customer loyalty scores, preferred channels, delivery completion timestamps, customer delivery ratings, and order booking channels. Cross-file relationship checks also showed that most operational records are fully linked, although some app-event records are not connected to order records, reflecting the fragmented and semi-structured data environment described in the case study.

The delivery-status inspection identified 616 on-time deliveries, 202 delayed deliveries, and 132 failed deliveries. This means that approximately 35.2% of delivery records were delayed or failed, providing early evidence of service-reliability issues that require deeper investigation in later SQL, R, and Python analysis.
